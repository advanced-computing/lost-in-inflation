```python
import pandas as pd
```

```python
inflation_data=pd.read_csv('/content/QuarterlyAnnualizedPercentChange-2025-q1.csv')
```

```python
inflation_data.head()
```

```python
#Kept getting errors with the dates format
inflation_data['Label'] = pd.to_datetime(inflation_data['Label'], errors='coerce')
inflation_data = inflation_data.set_index('Label')
```

```python
#Creating a simple time series
inflation_data[['CPI Inflation', 'Core CPI Inflation', 'PCE Inflation', 'Core PCE Inflation']].plot()

import matplotlib.pyplot as plt
plt.xlabel("Date")
plt.ylabel("Inflation Rate in %")

plt.title("U.S. Daily Inflation Forecast")

plt.show()
```
